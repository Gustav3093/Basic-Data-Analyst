# Positive ingredients for the human body.

This area will show which ingredient is beneficial to the human body and the appropriate amount, we will call this column (Suplements) where we will multiply 3 columns (Fixed Acidity, Residual Sugar, Free sulfur dioxide).


```python
df = pd.read_csv("/content/WineQuality.csv")
df['Suplements'] = df['fixed acidity'] * df['residual sugar']* df['free sulfur dioxide']
df.head(16)
```

# Alcohol Quatinty

In this area we will show from highest to lowest which types of wine contain more alcohol and which contain less.

```python
df['Highermag'] = df['alcohol'] > 10
df.head(16)
```

# Sugar level per wine
This area will show the sugar level per wine in the list.

```python
def sugar_level(level):
  if level <= 10:
    return 'Low sugar'
  elif level <= 20:
    return 'Medium sugar'
  else:
    return 'High sugar'

df['sugar_level'] = df['residual sugar'].apply(sugar_level)
df.head(16)
```

# Wine quality by sugar level

Here we will analyse the average quality of the wine by the type of sugar it contains.

```python
# Create a pivot table to analyze the average quality of wine by type and sugar level
pivot_table = pd.pivot_table(df, values='quality', index='Type', columns='sugar_level', aggfunc='mean')
# Display the pivot table
pivot_table
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Average Quality of Wine by Type and Sugar Level')
plt.show()
```

# Wine Quality by Alcohol Level

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='quality', y='alcohol', color = 'red', data=df)
plt.title('Average quality of wine by alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```
